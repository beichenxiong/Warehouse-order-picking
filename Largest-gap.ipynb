# Largest gap heuristic

* 引入本项目所需程序包

```python
import numpy as np
import timeit
```

* 读取picker location数据

```python
# loadtxt, 明确指定每列数据的类型
filename = './picker location.txt'
picker_location = np.loadtxt(filename,  #文件名
                        dtype=int)      #数据类型
print picker_location, picker_location.shape
```

* 读取包含前侧和后侧决策点的所有位置数据

```python
# loadtxt, 明确指定每列数据的类型
filename = './AllLocation.csv'
All_location = np.loadtxt(filename,         #文件名
                          delimiter=',',    #分割符
                          dtype=int         #数据类型
                         )  
print All_location, All_location.shape
```

* 将位置数据按照block，subaisle和picker的顺序进行排序

```python
idex=np.lexsort([All_location[:,2],All_location[:,1],-1*All_location[:,0]])
a_sort=All_location[idex,:]
```

```python
idex1=np.lexsort([picker_location[:,2],picker_location[:,1],-1*picker_location[:,0]])
a_sort1=picker_location[idex1,:]
```

* 首先计算每个点的位置坐标

```python
arr_coordition=np.zeros((244,4))
arr_coordition[:,0]=a_sort[:,0]
arr_coordition[:,1]=a_sort[:,1]
arr_coordition[:,2]=a_sort[:,2]
for i in range(0,122):
    if a_sort[i,2]==0:
        arr_coordition[i,3]=0
    elif a_sort[i,2]==13:
        arr_coordition[i,3]=27
    else:
        arr_coordition[i,3]=(a_sort[i,2]-1)*2+2.5
for i in range(122,244):
    if a_sort[i,2]==0:
        arr_coordition[i,3]=0
    elif a_sort[i,2]==11:
        arr_coordition[i,3]=23
    else:
        arr_coordition[i,3]=(a_sort[i,2]-1)*2+2.5
print arr_coordition
```

* 然后计算相邻拣货点之间，以及前后决策点与拣货点之间的距离

```python
arr_distance=np.zeros((244,4))
arr_distance[:,0]=a_sort[:,0]
arr_distance[:,1]=a_sort[:,1]
arr_distance[:,2]=a_sort[:,2]
for i in range(1,244):
    arr_distance[i,3]=arr_coordition[i,3]-arr_coordition[i-1,3]
print arr_distance
```

* 按照不同block将拣货点划分到数组中

```python
arr1_distance=arr_distance[arr_distance[:,0]==1]
arr2_distance=arr_distance[arr_distance[:,0]==2]
arr3_distance=arr_distance[arr_distance[:,0]==3]
arr4_distance=arr_distance[arr_distance[:,0]==4]
arr11=a_sort1[a_sort1[:,0]==1]
arr21=a_sort1[a_sort1[:,0]==2]
arr31=a_sort1[a_sort1[:,0]==3]
arr41=a_sort1[a_sort1[:,0]==4]
print arr1_distance,len(arr1_distance)
print '----------------------------'
print arr2_distance,len(arr2_distance)
print '----------------------------'
print arr3_distance,len(arr3_distance)
print '----------------------------'
print arr4_distance,len(arr4_distance)
```

* 接下来判断最远的需要拣货的block是block 4, 所以从block4开始优化

```python
len4=103
Forward_Route=[[1,1,0], #存储从block后侧的拣货路径
       [1,1,11],
       [2,1,0],
       [2,1,11],
      [3,1,0],
      [3,1,13],
      [4,1,0],
      [4,1,13],
      [4,2,13]]
Re_Route=np.zeros((1,3))  #存储从block前侧的拣货路径
Last_array=arr41[-1,1]   #判断最后一个需要拣货的走道
for i in range(2,Last_array):       
    arr4_sorted=arr4_distance[arr4_distance[:,1]==i]
    mix_distance4=np.max(arr4_sorted[:,3])     #选取每一个asile所有间隔中最大的间隔
    re=np.where(arr4_sorted==mix_distance4)    #选取最大间隔所在的位置
    mix_picker4=re[0][0]
    if mix_distance4==27:        #如果最大值为27，说明该通道内没有货物需要拣
        len4=len4+6              
        Forward_Route=np.vstack((Forward_Route,[4,i+1,13]))
        Re_Route=np.vstack((Re_Route,[4,i+1,0]))
    elif (arr4_sorted[mix_picker4,2]==13)&(mix_distance4<27): #如果最大值在后端决策点和相邻的拣货点之间，那么就从对面前端取货
        len4=len4+3+3+(27-mix_distance4)*2
        Forward_Route=np.vstack((Forward_Route,[4,i+1,13]))
        Re_Route=np.vstack((Re_Route,[4,i,arr4_sorted[mix_picker4-1,2]],[4,i,0],[4,i+1,0]))
    elif (arr4_sorted[mix_picker4-1,2]==0):       #如果最大间隔在前端决策点和相邻的拣货点之间，那么就从后端拣货
        len4=len4+3+3+2*(27-mix_distance4)
        Forward_Route=np.vstack((Forward_Route,[4,i,arr4_sorted[mix_picker4,2]],[4,i,13],[4,i+1,13]))
        Re_Route=np.vstack((Re_Route,[4,i+1,0]))
    else:
        len4=len4+3+3+2*(27-mix_distance4)        #如果最大间隔在相邻两个拣货点之间，那么从两端入口分别拣货
        Forward_Route=np.vstack((Forward_Route,[4,i,arr4_sorted[mix_picker4,2]],[4,i,13],[4,i+1,13]))
        Re_Route=np.vstack((Re_Route,[4,i,arr4_sorted[mix_picker4-1,2]],[4,i,0],[4,i+1,0]))
len4=len4+24
Re_Route[0,:]=[4,2,0]
idex4=np.lexsort([-1*Re_Route[:,1]])
Re_Route=Re_Route[idex4,:]
Route4=np.vstack((Forward_Route,Re_Route))      #后侧拣货路径与前侧拣货路径合起来就是block的最优拣货路径
print len4, Route4
```

* 同样的方法对于block3进行优化

```python
len3=0
Forward_Route=np.zeros((1,3))
Re_Route=np.zeros((1,3))
Last_array=arr31[-1,1]   #判断最后一个需要拣货的走道
for i in range(2,Last_array):       
    arr3_sorted=arr3_distance[arr3_distance[:,1]==i]
    mix_distance3=np.max(arr3_sorted[:,3])
    re=np.where(arr3_sorted==mix_distance3)    #选取最大间隔所在的位置
    mix_picker3=re[0][0]
    if mix_distance3==27:        #如果最大值为27，说明该通道内没有货物需要拣
        len3=len3+6              
        Forward_Route=np.vstack((Forward_Route,[3,i+1,13]))
        Re_Route=np.vstack((Re_Route,[3,i+1,0]))
    elif (arr3_sorted[mix_picker3,2]==13)&(mix_distance3<27):      #如果最大值在后端决策点和相邻的拣货点之间，那么就从对面取货
        len3=len3+3+3+(27-mix_distance3)*2
        Forward_Route=np.vstack((Forward_Route,[3,i+1,13]))
        Re_Route=np.vstack((Re_Route,[3,i,arr3_sorted[mix_picker3-1,2]],[3,i,0],[3,i+1,0]))
    elif arr3_sorted[mix_picker3-1,2]==0:
        len3=len3+3+3+2*(27-mix_distance3)
        Forward_Route=np.vstack((Forward_Route,[3,i,arr3_sorted[mix_picker3,2]],[3,i,13],[3,i+1,13]))
        Re_Route=np.vstack((Re_Route,[3,i+1,0]))
    else:
        len3=len3+3+3+2*(27-mix_distance3)
        Forward_Route=np.vstack((Forward_Route,[3,i,arr3_sorted[mix_picker3,2]],[3,i,13],[3,i+1,13]))
        Re_Route=np.vstack((Re_Route,[3,i,arr3_sorted[mix_picker3-1,2]],[3,i,0],[3,i+1,0]))
len3=len3+27
idex3=np.lexsort([-1*Re_Route[:,1]])
Re_Route=Re_Route[idex3,:]
Forward_Route[0,:]=[3,2,13]
Re_Route[-1,:]=[3,2,0]
Route3=np.vstack((Forward_Route,Re_Route))
print len3, Route3
```

* 同样的方法对于block3进行优化

```python
len2=0
Forward_Route=np.zeros((1,3))
Re_Route=np.zeros((1,3))
Last_array=arr21[-1,1]   #判断最后一个需要拣货的走道
for i in range(2,Last_array):       
    arr2_sorted=arr2_distance[arr2_distance[:,1]==i]
    mix_distance2=np.max(arr2_sorted[:,3])
    re=np.where(arr2_sorted==mix_distance2)    #选取最大间隔所在的位置
    mix_picker2=re[0][0]
    if mix_distance2==23:        #如果最大值为27，说明该通道内没有货物需要拣
        len2=len2+6              
        Forward_Route=np.vstack((Forward_Route,[2,i+1,11]))
        Re_Route=np.vstack((Re_Route,[2,i+1,0]))
    elif (arr2_sorted[mix_picker2,2]==11)&(mix_distance2<23):      #如果最大值在后端决策点和相邻的拣货点之间，那么就从对面取货
        len2=len2+3+3+(23-mix_distance2)*2
        Forward_Route=np.vstack((Forward_Route,[2,i+1,11]))
        Re_Route=np.vstack((Re_Route,[2,i,arr2_sorted[mix_picker2-1,2]],[2,i,0],[2,i+1,0]))
    elif arr2_sorted[mix_picker2-1,2]==0:
        len2=len2+3+3+2*(23-mix_distance2)
        Forward_Route=np.vstack((Forward_Route,[2,i,arr2_sorted[mix_picker2,2]],[2,i,11],[2,i+1,11]))
        Re_Route=np.vstack((Re_Route,[2,i+1,0]))
    else:
        len2=len2+3+3+2*(23-mix_distance2)
        Forward_Route=np.vstack((Forward_Route,[2,i,arr2_sorted[mix_picker2,2]],[2,i,11],[2,i+1,11]))
        Re_Route=np.vstack((Re_Route,[2,i,arr2_sorted[mix_picker2-1,2]],[2,i,0],[2,i+1,0]))
len2=len2+23
idex2=np.lexsort([-1*Re_Route[:,1]])
Re_Route=Re_Route[idex2,:]
Forward_Route[0,:]=[2,2,11]
Re_Route[-1,:]=[2,2,0]
Route2=np.vstack((Forward_Route,Re_Route))
print len2, Route2
```

* 同样的方法对于block1进行优化

```python
len1=0
Forward_Route=np.zeros((1,3))
Re_Route=np.zeros((1,3))
Last_array=arr11[-1,1]   #判断最后一个需要拣货的走道
for i in range(2,Last_array):       
    arr1_sorted=arr1_distance[arr1_distance[:,1]==i]
    mix_distance1=np.max(arr1_sorted[:,3])
    re=np.where(arr1_sorted==mix_distance1)    #选取最大间隔所在的位置
    mix_picker1=re[0][0]
    if mix_distance1==23:        #如果最大值为27，说明该通道内没有货物需要拣
        len1=len1+6              
        Forward_Route=np.vstack((Forward_Route,[1,i+1,11]))
        Re_Route=np.vstack((Re_Route,[1,i+1,0]))
    elif (arr1_sorted[mix_picker1,2]==11)&(mix_distance1<23):      #如果最大值在后端决策点和相邻的拣货点之间，那么就从对面取货
        len1=len1+3+3+(23-mix_distance1)*2
        Forward_Route=np.vstack((Forward_Route,[1,i+1,11]))
        Re_Route=np.vstack((Re_Route,[1,i,arr1_sorted[mix_picker1-1,2]],[1,i,0],[1,i+1,0]))
    elif arr1_sorted[mix_picker1-1,2]==0:
        len1=len1+3+3+2*(23-mix_distance1)
        Forward_Route=np.vstack((Forward_Route,[1,i,arr1_sorted[mix_picker1,2]],[1,i,11],[1,i+1,11]))
        Re_Route=np.vstack((Re_Route,[1,i+1,0]))
    else:
        len1=len1+3+3+2*(23-mix_distance1)
        Forward_Route=np.vstack((Forward_Route,[1,i,arr1_sorted[mix_picker1,2]],[1,i,11],[1,i+1,11]))
        Re_Route=np.vstack((Re_Route,[1,i,arr1_sorted[mix_picker1-1,2]],[1,i,0],[1,i+1,0]))
len1=len1+26
idex1=np.lexsort([-1*Re_Route[:,1]])
Re_Route=Re_Route[idex1,:]
Forward_Route[0,:]=[1,2,11]
Re_Route[-1,:]=[1,2,0]
Re_Route=np.vstack((Re_Route,[1,1,0]))
Route1=np.vstack((Forward_Route,Re_Route))
print len1, Route1  
```
