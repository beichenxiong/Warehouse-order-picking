# Combined Heuristic

* 引入本项目所需程序包

```python
import numpy as np
import timeit
```

* 读取picker location数据

```python
# loadtxt, 明确指定每列数据的类型
filename = './picker location.txt'
picker_location = np.loadtxt(filename,  #文件名
                        dtype=int)      #数据类型
print picker_location, picker_location.shape
```

* 读取包含前侧和后侧决策点的所有位置数据

```python
# loadtxt, 明确指定每列数据的类型
filename = './AllLocation.csv'
All_location = np.loadtxt(filename,         #文件名
                          delimiter=',',    #分割符
                          dtype=int         #数据类型
                         )  
print All_location, All_location.shape
```

* 将位置数据按照block，subaisle和picker的顺序进行排序

```python
idex=np.lexsort([All_location[:,2],All_location[:,1],-1*All_location[:,0]])
a_sort=All_location[idex,:]
```

```python
idex1=np.lexsort([picker_location[:,2],picker_location[:,1],-1*picker_location[:,0]])
a_sort1=picker_location[idex1,:]
```

* 首先计算每个点的位置坐标

```python
arr_coordition=np.zeros((244,4))
arr_coordition[:,0]=a_sort[:,0]
arr_coordition[:,1]=a_sort[:,1]
arr_coordition[:,2]=a_sort[:,2]
for i in range(0,122):
    if a_sort[i,2]==0:
        arr_coordition[i,3]=0
    elif a_sort[i,2]==13:
        arr_coordition[i,3]=27
    else:
        arr_coordition[i,3]=(a_sort[i,2]-1)*2+2.5
for i in range(122,244):
    if a_sort[i,2]==0:
        arr_coordition[i,3]=0
    elif a_sort[i,2]==11:
        arr_coordition[i,3]=23
    else:
        arr_coordition[i,3]=(a_sort[i,2]-1)*2+2.5
print arr_coordition
```

* 按照不同block将拣货点划分到数组中

```python
arr1_coordition=arr_coordition[arr_coordition[:,0]==1]
arr2_coordition=arr_coordition[arr_coordition[:,0]==2]
arr3_coordition=arr_coordition[arr_coordition[:,0]==3]
arr4_coordition=arr_coordition[arr_coordition[:,0]==4]
arr11=a_sort1[a_sort1[:,0]==1]
arr21=a_sort1[a_sort1[:,0]==2]
arr31=a_sort1[a_sort1[:,0]==3]
arr41=a_sort1[a_sort1[:,0]==4]
print arr1_coordition,len(arr1_coordition)
print '----------------------------'
print arr2_coordition,len(arr2_coordition)
print '----------------------------'
print arr3_coordition,len(arr3_coordition)
print '----------------------------'
print arr4_coordition,len(arr4_coordition)
```

* 接下来判断最远的需要拣货的block是block 4, 所以从block4开始优化

```python
alen4=100
blen4=84
La_Route=[[1,1,0],
          [1,1,11],
          [2,1,0],
          [2,1,11],
          [3,1,0],
          [3,1,13],
          [4,1,0],
         [4,1,13]]
Lb_Route=[[1,1,0],
          [1,1,11],
          [2,1,0],
          [2,1,11],
          [3,1,0],
          [3,1,13],
          [4,1,0],
         [4,1,5],
         [4,1,0]]
Last_array=arr41[-1,1]   #判断最后一个需要拣货的走道
for i in range(2,Last_array):       
    arr4_sorted=arr4_coordition[arr4_coordition[:,1]==i]
    La1_Route=La_Route
    if len(arr4_sorted)==2:                       #当该巷道没有货的时候，就不用拣货，直接走到下一巷道
        alen4=alen4+3
        La_Route=np.vstack((La_Route,[4,i,13]))
        blen4=blen4+3
        Lb_Route=np.vstack((Lb_Route,[4,i,0]))
    else:                  
        Laa=alen4+3+2*(27- arr4_sorted[1,3])    #计算4个距离
        Lba=blen4+3+27
        Lab=alen4+3+27
        Lbb=blen4+3+2*arr4_sorted[-2,3]
        if Laa<Lba:                             #对于下一个点后端决策点来说，如果Laa<Lba说明从后端取货再返回后端距离比较短
            La_Route=np.vstack((La_Route,[4,i,13],[4,i,arr4_sorted[1,2]],[4,i,13]))
            alen4=Laa 
        else:                                   #否则是从前端穿过后端距离比较短  
            La_Route=np.vstack((Lb_Route,[4,i,0],[4,i,13]))
            alen4=Lba
        if Lbb<Lab:                             #对于下一个点前端决策点来说，如果Lbb<Lab说明从前端取货再返回前端距离比较短
            Lb_Route=np.vstack((Lb_Route,[4,i,0],[4,i,arr4_sorted[-2,2]],[4,i,0]))
            blen4=Lbb
        else:                                   #否则是从前端穿过后端距离比较短  
            Lb_Route=np.vstack((La1_Route,[4,i,13],[4,i,0]))
            blen4=Lab
Lab=alen4+3+27                                  #计算在最右拣货通道的距离，Lab是从后端穿越到前端，Lbb是从前端取货再返回前端
Lbb=blen4+3+2*arr4_coordition[-4,3]
if Lbb<Lab:                                     #说明从前端取货再返回前端的距离比较短
    blen4=Lbb
    Route4=np.vstack((Lb_Route,[4,arr4_coordition[-4,1],0],[4,arr4_coordition[-4,1],arr4_coordition[-4,2]],[4,arr4_coordition[-4,1],0]))
else:                                           #否则应该从后端穿越到前端
    Route4=np.vstack((La_Route,[4,arr4_coordition[-4,1],13],[4,arr4_coordition[-4,1],0]))
    blen4=Lab
print blen4, Route4
```

* 同样的方法对于block3进行优化

```python
alen3=8
blen3=30
La_Route=[[3,17,13],
         [3,18,13],
         [3,18,12],
         [3,18,13]]
Lb_Route=[[3,17,13],
         [3,18,13],
         [3,18,0]]
Right_array=arr31[-1,1]   #判断最后一个需要拣货的走道
Left_array=arr31[2,1]     #
for i in range(Right_array-1,2,-1):       
    arr3_sorted=arr3_coordition[arr3_coordition[:,1]==i]
    La1_Route=La_Route
    if len(arr3_sorted)==2:                       #当该巷道没有货的时候，就不用拣货，直接走到下一巷道
        alen3=alen3+3
        La_Route=np.vstack((La_Route,[3,i,13]))
        blen3=blen3+3
        Lb_Route=np.vstack((Lb_Route,[3,i,0]))
    else:                  
        Laa=alen3+3+2*(27- arr3_sorted[1,3])    #计算4个距离
        Lba=blen3+3+27
        Lab=alen3+3+27
        Lbb=blen3+3+2*arr3_sorted[-2,3]
        if Laa<Lba:                             #对于下一个点后端决策点来说，有两种方法到达，一种是从后端取货再返回后端，
            La_Route=np.vstack((La_Route,[3,i,13],[3,i,arr3_sorted[1,2]],[3,i,13]))
            alen3=Laa 
        else:                                   #另一种是从前端穿过后端    
            La_Route=np.vstack((Lb_Route,[3,i,0],[3,i,13]))
            alen3=Lba
        if Lbb<Lab:
            Lb_Route=np.vstack((Lb_Route,[3,i,0],[3,i,arr3_sorted[-2,2]],[3,i,0]))
            blen3=Lbb
        else:
            Lb_Route=np.vstack((La1_Route,[3,i,13],[3,i,0]))
            blen3=Lab
Lab=alen3+3+27
Lbb=blen3+3+2*arr3_coordition[5,3]
if Lbb<Lab:
    blen3=Lbb
    Route3=np.vstack((Lb_Route,[3,arr3_coordition[5,1],0],[3,arr3_coordition[5,1],arr3_coordition[5,2]],[3,arr3_coordition[5,1],0]))
else:
    Route3=np.vstack((La_Route,[3,arr3_coordition[5,1],13],[3,arr3_coordition[5,1],0]))
    blen3=Lab
print blen3, Route3
```

* 同样的方法对于block2进行优化

```python
alen2=44
blen2=26
La_Route=[[2,2,11],
         [2,3,11],
         [2,3,1],
         [2,3,11]]
Lb_Route=[[2,2,11],
         [2,3,11],
         [2,3,0]]
Right_array=arr21[-1,1]   #判断最后一个需要拣货的走道
for i in range(2,Right_array):       
    arr2_sorted=arr2_coordition[arr2_coordition[:,1]==i]
    La1_Route=La_Route
    if len(arr2_sorted)==2:                       #当该巷道没有货的时候，就不用拣货，直接走到下一巷道
        alen2=alen2+3
        La_Route=np.vstack((La_Route,[2,i,11]))
        blen2=blen2+3
        Lb_Route=np.vstack((Lb_Route,[2,i,0]))
    else:                  
        Laa=alen2+3+2*(23- arr2_sorted[1,3])    #计算4个距离
        Lba=blen2+3+23
        Lab=alen2+3+23
        Lbb=blen2+3+2*arr2_sorted[-2,3]
        if Laa<Lba:                             #对于下一个点后端决策点来说，有两种方法到达，一种是从后端取货再返回后端，
            La_Route=np.vstack((La_Route,[2,i,11],[2,i,arr2_sorted[1,2]],[2,i,11]))
            alen2=Laa 
        else:                                   #另一种是从前端穿过后端    
            La_Route=np.vstack((Lb_Route,[2,i,0],[2,i,11]))
            alen2=Lba
        if Lbb<Lab:
            Lb_Route=np.vstack((Lb_Route,[2,i,0],[2,i,arr2_sorted[-2,2]],[2,i,0]))
            blen2=Lbb
        else:
            Lb_Route=np.vstack((La1_Route,[2,i,11],[2,i,0]))
            blen2=Lab
Lab=alen2+3+23
Lbb=blen2+3+2*arr2_coordition[-4,3]
if Lbb<Lab:
    blen2=Lbb
    Route2=np.vstack((Lb_Route,[2,arr2_coordition[-4,1],0],[2,arr2_coordition[-4,1],arr2_coordition[-4,2]],[2,arr2_coordition[-4,1],0]))
else:
    Route2=np.vstack((La_Route,[2,arr2_coordition[-4,1],13],[2,arr2_coordition[-4,1],0]))
    blen2=Lab
print blen2, Route2
```

* 同样的方法对于block1进行优化

```python
alen1=13
blen1=23
La_Route=[[2,17,13],
         [2,17,8],
         [2,17,13]]
Lb_Route=[[2,17,13],
         [2,17,0]]
Right_array=arr11[-1,1]   #判断最后一个需要拣货的走道
Left_array=arr11[1,1]
for i in range(Right_array-1,3,-1):       
    arr1_sorted=arr1_coordition[arr1_coordition[:,1]==i]
    La1_Route=La_Route
    if len(arr1_sorted)==2:                       #当该巷道没有货的时候，就不用拣货，直接走到下一巷道
        alen1=alen1+3
        La_Route=np.vstack((La_Route,[1,i,13]))
        blen1=blen1+3
        Lb_Route=np.vstack((Lb_Route,[1,i,0]))
    else:                  
        Laa=alen1+3+2*(23- arr1_sorted[1,3])    #计算4个距离
        Lba=blen1+3+23
        Lab=alen1+3+23
        Lbb=blen1+3+2*arr1_sorted[-2,3]
        if Laa<Lba:                             #对于下一个点后端决策点来说，有两种方法到达，一种是从后端取货再返回后端，
            La_Route=np.vstack((La_Route,[1,i,11],[1,i,arr1_sorted[1,2]],[1,i,11]))
            alen1=Laa 
        else:                                   #另一种是从前端穿过后端    
            La_Route=np.vstack((Lb_Route,[1,i,0],[1,i,11]))
            alen1=Lba
        if Lbb<Lab:
            Lb_Route=np.vstack((Lb_Route,[1,i,0],[1,i,arr1_sorted[-2,2]],[1,i,0]))
            blen1=Lbb
        else:
            Lb_Route=np.vstack((La1_Route,[1,i,11],[1,i,0]))
            blen1=Lab
Lab=alen1+3+23
Lbb=blen1+3+2*arr1_coordition[6,3]
if Lbb<Lab:
    blen1=Lbb
    Route1=np.vstack((Lb_Route,[1,arr1_coordition[6,1],0],[1,arr1_coordition[6,1],arr1_coordition[6,2]],[1,arr1_coordition[6,1],0]))
else:
    Route1=np.vstack((La_Route,[1,arr1_coordition[6,1],11],[1,arr1_coordition[6,1],0]))
    blen1=Lab
Route1=np.vstack((Route1,[1,2,0],[1,1,0]))
blen1=blen1+6
print blen1, Route1
```
