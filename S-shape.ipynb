```python
idex1=np.lexsort([picker_location[:,2],picker_location[:,1],-1*picker_location[:,0]])
a_sort1=picker_location[idex1,:]
```

* 使用切片功能，将每一个block的位置信息存储到数组中

```python
arr1=a_sort[a_sort[:,0]==1]
arr2=a_sort[a_sort[:,0]==2]
arr3=a_sort[a_sort[:,0]==3]
arr4=a_sort[a_sort[:,0]==4]
arr11=a_sort1[a_sort1[:,0]==1]
arr21=a_sort1[a_sort1[:,0]==2]
arr31=a_sort1[a_sort1[:,0]==3]
arr41=a_sort1[a_sort1[:,0]==4]
print arr1
print '-------------'
print arr2
print '-------------'
print arr3
print '-------------'
print arr4
```

* 首先从最远的block4开始拣货，计算拣货路径和距离

```python
len4=73            #首先穿过整个aisle到达最远的block的前面
tra4=0
Route=[[1,1,0],
       [1,1,11],
       [2,1,0],
       [2,1,11],
      [3,1,0],
      [3,1,13]]
Last_array=arr41[-1,1]   #判断最后一个需要拣货的走道
Last_picker=arr41[-1,2]  #判断最后一个需要拣货通道中最远的货位
for i in range(1,Last_array):       #从第一个通道开始拣货
    if len(arr4[arr4[:,1]==i])>2:   #判断该通道是否需要拣货
        len4=len4+30                #如果需要那么就在总距离上加30
        tra4=tra4+1                 #tra4帮助判断拣货的方向是从前往后，还是从后往前
        if tra4%2==1:               #如果tra4是奇数，那么就从前往后拣货
            Route=np.r_[Route,arr4[arr4[:,1]==i]]
        else:                       #否则，从后往前取拣货
            arr4_sort=arr4[arr4[:,1]==i]
            idex4=np.lexsort([-1*arr4_sort[:,2]])
            Rearr4_sort=arr4_sort[idex4,:]
            Route=np.r_[Route,Rearr4_sort]
    else:                           #如果该通道不需要拣货
        len4=len4+3                 #那么就直走，不进入通道，总距离加3
        Next_picker=Route[-1,:]
        Next_picker[1]=Next_picker[1]+1
        Route=np.vstack((Route,Next_picker))
if Route[-1,2]==0:                  #到最后一个需要拣货的通道，如果是在block的前面，那么就到达最远拣货位然后再返回
    len4=len4+(Last_picker-1)*4+5
    Last_Route=arr4[arr4[:,1]==Last_array]
    Last_Route[-1,2]=0
    Route=np.r_[Route,Last_Route]
else:                              #如果是在block的后面，那么直接穿过该通道，总距离加27
    len4=len4+27
    Last_Route=arr4[arr4[:,1]==Last_array]
    idex41=np.lexsort([-1*arr4_sort[:,2]])
    ReLast_Route=Last_Route[idex41,:]
    Route=np.r_[Route,ReLast_Route]
print len4, Route
```

* 接着我们计算block3的拣货路径和距离，注意是从最右边的需要拣货的通道开始拣货

```python
len3=3
tra3=0
First_array=arr31[-1,1]    #判断最右边的需要拣货的通道
Last_picker=arr31[2,2]
for i in range(First_array,2,-1):
    if len(arr3[arr3[:,1]==i])>2:
        len3=len3+30
        tra3=tra3+1
        if tra3%2==1:
            arr3_sort=arr3[arr3[:,1]==i]
            idex3=np.lexsort([-1*arr3_sort[:,2]])
            Rearr3_sort=arr3_sort[idex3,:]
            Route=np.r_[Route,Rearr3_sort]
        else:
            Route=np.r_[Route,arr3[arr3[:,1]==i]]
    else:
        len3=len3+3
        Next_picker=Route[-1,:]
        Next_picker[1]+=1
        Route=np.vstack((Route,Next_picker))
if Route[-1,2]==0:
    len3=len3+(Last_picker-1)*4+5
    Last_Route=arr3[arr3[:,1]==2]
    Last_Route[-1,2]=0
    Route=np.r_[Route,Last_Route]
else:
    len3=len3+27
    Last_Route=arr3[arr3[:,1]==2]
    idex31=np.lexsort([-1*arr3_sort[:,2]])
    ReLast_Route=Last_Route[idex31,:]
    Route=np.r_[Route,ReLast_Route]
print Route,len3
```

* 接着我们计算block2的拣货路径和距离

```python
len2=3
tra2=0
Last_array=arr21[-1,1]
Last_picker=arr21[-1,2]
First_array=Route[-1,:]
First_array[0]-=1
First_array[2]=11
Route=np.row_stack((Route,First_array))
for i in range(3,Last_array):
    if len(arr2[arr2[:,1]==i])>2:
        len2=len2+26
        tra2=tra2+1
        if tra2%2==0:
            Route=np.r_[Route,arr2[arr2[:,1]==i]]
        else:
            arr2_sort=arr2[arr2[:,1]==i]
            idex2=np.lexsort([-1*arr2_sort[:,2]])
            Rearr2_sort=arr2_sort[idex2,:]
            Route=np.r_[Route,Rearr2_sort]
    else:
        len2=len2+3
        Next_picker=Route[-1,:]
        Next_picker[1]+=1
        Route=np.vstack((Route,Next_picker))
if Route[-1,2]==0:
    len2=len2+(Last_picker-1)*4+5
    Last_Route=arr2[arr2[:,1]==Last_array]
    Last_Route[-1,2]=0
    Route=np.r_[Route,Last_Route]
else:
    len2=len2+23
    Last_Route=arr2[arr2[:,1]==Last_array]
    idex21=np.lexsort([-1*arr2_sort[:,2]])
    ReLast_Route=Last_Route[idex21,:]
    Route=np.r_[Route,ReLast_Route]
print len2,Route
```

```python
len1=6
tra1=0
First_array=arr11[-1,1]
Last_picker=arr11[4,2]
for i in range(First_array,3,-1):
    if len(arr1[arr1[:,1]==i])>2:
        len1=len1+26
        tra1=tra1+1
        if tra1%2==1:
            arr1_sort=arr1[arr1[:,1]==i]
            idex1=np.lexsort([-1*arr1_sort[:,2]])
            Rearr1_sort=arr1_sort[idex1,:]
            Route=np.r_[Route,Rearr1_sort]
        else:
            Route=np.r_[Route,arr1[arr1[:,1]==i]]
    else:
        len1=len1+3
        Next_picker=Route[-1,:]
        Next_picker[1]+=1
        Route=np.row_stack((Route,Next_picker))
if Route[-1,2]==0:
    len1=len1+(Last_picker-1)*4+5
    Last_Route=arr1[arr1[:,1]==3]
    Last_Route[-1,2]=0
    Route=np.r_[Route,Last_Route]
else:
    len1=len1+23
    Last_Route=arr1[arr1[:,1]==3]
    idex11=np.lexsort([-1*arr1_sort[:,2]])
    ReLast_Route=Last_Route[idex11,:]
    Route=np.r_[Route,ReLast_Route]
Leave_route=[[1,2,0],
            [1,1,0]]
Route=np.r_[Route,Leave_route]
print Route,len1
```
